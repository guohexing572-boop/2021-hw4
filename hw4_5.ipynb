```python
import torch.nn.functional as F
class LightweightConformerBlock(nn.Module):
    def __init__(self, d_model=80, conv_kernel_size=31, dropout=0.1):
        super().__init__()
        
        # å‰é¦ˆç½‘ç»œéƒ¨åˆ† (ç®€åŒ–ç‰ˆ)
        self.ffn = nn.Sequential(
            nn.Linear(d_model, d_model * 2),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(d_model * 2, d_model),
            nn.Dropout(dropout)
        )
        self.norm1 = nn.LayerNorm(d_model)
        
        # åŒå¤´è‡ªæ³¨æ„åŠ› (ä¿®æ”¹è¿™é‡Œ: num_heads=2)
        self.self_attn = nn.MultiheadAttention(
            d_model, num_heads=2, dropout=dropout, batch_first=True  # num_heads=2
        )
        self.norm2 = nn.LayerNorm(d_model)
        
        # è½»é‡çº§å·ç§¯æ¨¡å—
        self.conv_module = LightweightConvModule(d_model, conv_kernel_size, dropout)
        self.norm3 = nn.LayerNorm(d_model)
        
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        # å‰é¦ˆç½‘ç»œ + æ®‹å·®
        residual = x
        x = self.norm1(x)
        x = self.ffn(x)
        x = self.dropout(x)
        x = residual + x
        
        # è‡ªæ³¨æ„åŠ› + æ®‹å·®
        residual = x
        x = self.norm2(x)
        x, _ = self.self_attn(x, x, x)
        x = self.dropout(x)
        x = residual + x
        
        # å·ç§¯æ¨¡å— + æ®‹å·®
        residual = x
        x = self.norm3(x)
        x = self.conv_module(x)
        x = self.dropout(x)
        x = residual + x
        
        return x

class LightweightConvModule(nn.Module):
    def __init__(self, d_model, kernel_size=31, dropout=0.1):
        super().__init__()
        assert kernel_size % 2 == 1, "Kernel size should be odd"
        
        # ç®€åŒ–ç‰ˆå·ç§¯æ¨¡å—
        self.pointwise_conv1 = nn.Conv1d(d_model, d_model, kernel_size=1)
        self.depthwise_conv = nn.Conv1d(
            d_model, d_model, kernel_size=kernel_size,
            padding=(kernel_size-1)//2, groups=d_model
        )
        self.batch_norm = nn.BatchNorm1d(d_model)
        self.activation = nn.ReLU()
        self.pointwise_conv2 = nn.Conv1d(d_model, d_model, kernel_size=1)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, x):
        # è¾“å…¥å½¢çŠ¶: (batch, seq_len, d_model)
        residual = x
        
        # è½¬ç½®ç”¨äºå·ç§¯
        x = x.transpose(1, 2)  # (batch, d_model, seq_len)
        
        # ç‚¹å¼å·ç§¯
        x = self.pointwise_conv1(x)
        
        # æ·±åº¦å¯åˆ†ç¦»å·ç§¯
        x = self.depthwise_conv(x)
        x = self.batch_norm(x)
        x = self.activation(x)
        
        # ç‚¹å¼å·ç§¯
        x = self.pointwise_conv2(x)
        x = self.dropout(x)
        
        # æ¢å¤ç»´åº¦
        x = x.transpose(1, 2)  # (batch, seq_len, d_model)
        
        return x + residual

class LightweightConformerClassifier(nn.Module):
    def __init__(self, d_model=80, n_spks=600, dropout=0.1, num_layers=2):
        """è½»é‡çº§Conformeråˆ†ç±»å™¨ï¼ˆå•å¤´æ³¨æ„åŠ›ï¼‰
        
        Args:
            d_model: ç‰¹å¾ç»´åº¦
            n_spks: è¯´è¯äººæ•°é‡
            dropout: dropoutç‡
            num_layers: Conformerå±‚æ•°
        """
        super().__init__()
        
        # è¾“å…¥æŠ•å½±å±‚
        self.prenet = nn.Linear(40, d_model)
        self.prenet_dropout = nn.Dropout(dropout)
        
        # Conformerå±‚ (åŒå¤´æ³¨æ„åŠ›)
        self.conformer_layers = nn.ModuleList([
            LightweightConformerBlock(
                d_model=d_model,
                conv_kernel_size=31,
                dropout=dropout
            ) for _ in range(num_layers)
        ])
        
        # è¾“å‡ºå±‚
        self.output_layer = nn.Sequential(
            nn.Linear(d_model, d_model),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(d_model, n_spks),
        )

    def forward(self, mels):
        """
        Args:
            mels: (batch_size, seq_len, 40)
        Returns:
            out: (batch_size, n_spks)
        """
        # è¾“å…¥æŠ•å½±
        x = self.prenet(mels)  # (batch, seq_len, d_model)
        x = self.prenet_dropout(x)
        
        # Conformerç¼–ç 
        for layer in self.conformer_layers:
            x = layer(x)
        
        # å…¨å±€å¹³å‡æ± åŒ–
        x = x.mean(dim=1)  # (batch, d_model)
        
        # åˆ†ç±»è¾“å‡º
        out = self.output_layer(x)  # (batch, n_spks)
        
        return out
```

```python
class Classifier(nn.Module):
  def __init__(self, d_model=80, n_spks=600, dropout=0.1):
    """åˆå§‹åŒ–åˆ†ç±»å™¨æ¨¡å‹
    
    Args:
        d_model (int): æ¨¡å‹çš„ç‰¹å¾ç»´åº¦ï¼Œé»˜è®¤ä¸º80
        n_spks (int): è¯´è¯äººæ•°é‡ï¼Œé»˜è®¤ä¸º600
        dropout (float): dropoutç‡ï¼Œé»˜è®¤ä¸º0.1
    """
    super().__init__()
    
    # å°†è¾“å…¥ç‰¹å¾çš„ç»´åº¦ä»40æŠ•å½±åˆ°d_model
    # è¾“å…¥: (batch_size, length, 40) -> è¾“å‡º: (batch_size, length, d_model)
    self.prenet = nn.Linear(40, d_model)
    
    # TODO: å°†Transformeræ”¹ä¸ºConformer
    # å‚è€ƒè®ºæ–‡: https://arxiv.org/abs/2005.08100
    # Conformerç»“åˆäº†CNNå’ŒTransformerçš„ä¼˜ç‚¹ï¼Œåœ¨è¯­éŸ³ä»»åŠ¡ä¸Šè¡¨ç°æ›´å¥½
    
    # å½“å‰ä½¿ç”¨Transformerç¼–ç å±‚
    self.encoder_layer = nn.TransformerEncoderLayer(
      d_model=d_model,        # ç‰¹å¾ç»´åº¦
      dim_feedforward=256,    # å‰é¦ˆç½‘ç»œçš„éšè—å±‚ç»´åº¦
      nhead=2                 # æ³¨æ„åŠ›å¤´æ•°
    )
    # å¦‚æœéœ€è¦å¤šå±‚ï¼Œå¯ä»¥ä½¿ç”¨TransformerEncoder
    # self.encoder = nn.TransformerEncoder(self.encoder_layer, num_layers=2)

    # é¢„æµ‹å±‚ï¼šå°†d_modelç»´ç‰¹å¾æ˜ å°„åˆ°è¯´è¯äººæ•°é‡
    self.pred_layer = nn.Sequential(
      nn.Linear(d_model, d_model),  # çº¿æ€§å˜æ¢
      nn.ReLU(),                    # æ¿€æ´»å‡½æ•°
      nn.Linear(d_model, n_spks),   # è¾“å‡ºå±‚ï¼Œè¾“å‡ºæ¯ä¸ªè¯´è¯äººçš„åˆ†æ•°
    )

  def forward(self, mels):
    """
    å‰å‘ä¼ æ’­
    
    Args:
      mels: è¾“å…¥melé¢‘è°±å›¾ï¼Œå½¢çŠ¶ä¸º (batch_size, length, 40)
      
    Return:
      out: è¾“å‡ºè¯´è¯äººåˆ†ç±»ç»“æœï¼Œå½¢çŠ¶ä¸º (batch_size, n_spks)
    """
    # è¾“å…¥æŠ•å½±: (batch_size, length, 40) -> (batch_size, length, d_model)
    out = self.prenet(mels)
    
    # è°ƒæ•´ç»´åº¦ä»¥é€‚åº”Transformerè¾“å…¥è¦æ±‚
    # (batch_size, length, d_model) -> (length, batch_size, d_model)
    out = out.permute(1, 0, 2)
    
    # Transformerç¼–ç å±‚æœŸæœ›è¾“å…¥å½¢çŠ¶ä¸º (length, batch_size, d_model)
    out = self.encoder_layer(out)
    
    # æ¢å¤ç»´åº¦: (length, batch_size, d_model) -> (batch_size, length, d_model)
    out = out.transpose(0, 1)
    
    # å‡å€¼æ± åŒ–ï¼šæ²¿æ—¶é—´ç»´åº¦æ±‚å¹³å‡
    # (batch_size, length, d_model) -> (batch_size, d_model)
    stats = out.mean(dim=1)

    # é€šè¿‡é¢„æµ‹å±‚å¾—åˆ°æœ€ç»ˆåˆ†ç±»ç»“æœ
    # (batch_size, d_model) -> (batch_size, n_spks)
    out = self.pred_layer(stats)
    
    return out
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ConformerBlock(nn.Module):
    def __init__(self, d_model, nhead, dim_feedforward, dropout=0.1, conv_kernel_size=31):
        super().__init__()
        self.d_model = d_model
        
        # ç¬¬ä¸€éƒ¨åˆ†ï¼šå‰é¦ˆç½‘ç»œ
        self.ffn1 = nn.Sequential(
            nn.Linear(d_model, dim_feedforward),
            nn.SiLU(),
            nn.Dropout(dropout),
            nn.Linear(dim_feedforward, d_model),
            nn.Dropout(dropout)
        )
        self.norm1 = nn.LayerNorm(d_model)
        
        # ç¬¬äºŒéƒ¨åˆ†ï¼šè‡ªæ³¨æ„åŠ›
        self.self_attn = nn.MultiheadAttention(d_model, nhead, dropout=dropout, batch_first=True)
        self.norm2 = nn.LayerNorm(d_model)
        
        # ç¬¬ä¸‰éƒ¨åˆ†ï¼šå·ç§¯æ¨¡å—
        self.conv_module = nn.Sequential(
            nn.LayerNorm(d_model),
            nn.Conv1d(d_model, 2*d_model, 1),  # ç‚¹å¼å·ç§¯
            nn.GLU(dim=1),  # GLUæ¿€æ´»
            nn.Conv1d(d_model, d_model, conv_kernel_size, 
                     padding=(conv_kernel_size-1)//2, groups=d_model),  # æ·±åº¦å¯åˆ†ç¦»å·ç§¯
            nn.BatchNorm1d(d_model),
            nn.SiLU(),
            nn.Conv1d(d_model, d_model, 1),  # ç‚¹å¼å·ç§¯
            nn.Dropout(dropout)
        )
        self.norm3 = nn.LayerNorm(d_model)
        
        # ç¬¬å››éƒ¨åˆ†ï¼šç¬¬äºŒéƒ¨åˆ†å‰é¦ˆç½‘ç»œ
        self.ffn2 = nn.Sequential(
            nn.Linear(d_model, dim_feedforward),
            nn.SiLU(),
            nn.Dropout(dropout),
            nn.Linear(dim_feedforward, d_model),
            nn.Dropout(dropout)
        )
        self.norm4 = nn.LayerNorm(d_model)
        
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        # ç¬¬ä¸€éƒ¨åˆ†ï¼šå‰é¦ˆç½‘ç»œ + æ®‹å·®è¿æ¥
        residual = x
        x = self.norm1(x)
        x = self.ffn1(x)
        x = self.dropout(x)
        x = x + residual
        
        # ç¬¬äºŒéƒ¨åˆ†ï¼šè‡ªæ³¨æ„åŠ› + æ®‹å·®è¿æ¥
        residual = x
        x = self.norm2(x)
        x, _ = self.self_attn(x, x, x)
        x = self.dropout(x)
        x = x + residual
        
        # ç¬¬ä¸‰éƒ¨åˆ†ï¼šå·ç§¯æ¨¡å— + æ®‹å·®è¿æ¥
        residual = x
        x = self.norm3(x)
        # è°ƒæ•´ç»´åº¦ç”¨äºå·ç§¯: (batch, seq_len, d_model) -> (batch, d_model, seq_len)
        x_conv = x.transpose(1, 2)
        x_conv = self.conv_module(x_conv)
        x_conv = x_conv.transpose(1, 2)  # æ¢å¤ç»´åº¦
        x = self.dropout(x_conv)
        x = x + residual
        
        # ç¬¬å››éƒ¨åˆ†ï¼šå‰é¦ˆç½‘ç»œ + æ®‹å·®è¿æ¥
        residual = x
        x = self.norm4(x)
        x = self.ffn2(x)
        x = self.dropout(x)
        x = x + residual
        
        return x

class ConformerClassifier(nn.Module):
    def __init__(self, d_model=80, n_spks=600, dropout=0.1, num_layers=4):
        super().__init__()
        
        # è¾“å…¥æŠ•å½±å±‚
        self.prenet = nn.Linear(40, d_model)
        
        # Conformerç¼–ç å±‚
        self.conformer_layers = nn.ModuleList([
            ConformerBlock(
                d_model=d_model,
                nhead=8,  # æ›´å¤šçš„æ³¨æ„åŠ›å¤´
                dim_feedforward=256,
                dropout=dropout,
                conv_kernel_size=31
            ) for _ in range(num_layers)
        ])
        
        # åˆ†ç±»å™¨
        self.classifier = nn.Sequential(
            nn.Linear(d_model, d_model),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(d_model, n_spks),
        )

    def forward(self, mels):
        # è¾“å…¥æŠ•å½±
        x = self.prenet(mels)  # (batch, seq_len, d_model)
        
        # Conformerç¼–ç 
        for layer in self.conformer_layers:
            x = layer(x)
        
        # å…¨å±€å¹³å‡æ± åŒ–
        x = x.mean(dim=1)  # (batch, d_model)
        
        # åˆ†ç±»
        out = self.classifier(x)  # (batch, n_spks)
        
        return out
```

```python
class SpeakerClassifier(nn.Module):
    def __init__(self, d_model=80, n_spks=600, dropout=0.1):
        """è¯´è¯äººåˆ†ç±»æ¨¡å‹
        
        Args:
            d_model (int): æ¨¡å‹ç‰¹å¾ç»´åº¦
            n_spks (int): è¯´è¯äººæ•°é‡
            dropout (float): dropoutæ¯”ç‡
        """
        super().__init__()
        
        # è¾“å…¥æŠ•å½±å±‚: 40ç»´melç‰¹å¾ -> d_modelç»´
        self.prenet = nn.Linear(40, d_model)
        
        # Transformerç¼–ç å±‚
        self.encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            dim_feedforward=256,
            nhead=2,
            dropout=dropout,
            batch_first=False  # Transformeréœ€è¦åºåˆ—ç»´åº¦åœ¨å‰
        )
        
        # åˆ†ç±»å™¨
        self.classifier = nn.Sequential(
            nn.Linear(d_model, d_model),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(d_model, n_spks),
        )

    def forward(self, mels):
        """å‰å‘ä¼ æ’­
        
        Args:
            mels: melé¢‘è°±å›¾, shape: (batch, seq_len, 40)
            
        Returns:
            è¯´è¯äººåˆ†ç±»logits, shape: (batch, n_spks)
        """
        # è¾“å…¥æŠ•å½±
        x = self.prenet(mels)  # (batch, seq_len, d_model)
        
        # è°ƒæ•´ç»´åº¦é€‚åº”Transformer: (seq_len, batch, d_model)
        x = x.transpose(0, 1)
        
        # Transformerç¼–ç 
        x = self.encoder_layer(x)  # (seq_len, batch, d_model)
        
        # æ¢å¤ç»´åº¦: (batch, seq_len, d_model)
        x = x.transpose(0, 1)
        
        # å…¨å±€å¹³å‡æ± åŒ–
        x = x.mean(dim=1)  # (batch, d_model)
        
        # åˆ†ç±»
        out = self.classifier(x)  # (batch, n_spks)
        
        return out
```

```python
def plot_loss_curves(train_losses, val_losses, save_path=None):
    """ç»˜åˆ¶è®­ç»ƒå’ŒéªŒè¯æŸå¤±æ›²çº¿"""
    plt.figure(figsize=(10, 6))
    plt.plot(train_losses, label='è®­ç»ƒæŸå¤±', linewidth=2)
    plt.plot(val_losses, label='éªŒè¯æŸå¤±', linewidth=2)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('è®­ç»ƒå’ŒéªŒè¯æŸå¤±æ›²çº¿')
    plt.legend()
    plt.grid(True, alpha=0.3)

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"å›¾ç‰‡å·²ä¿å­˜åˆ°: {save_path}")
```

```python
def plot_accuracy_curves(train_accuracies, val_accuracies, save_path=None):
    """ç»˜åˆ¶å‡†ç¡®ç‡æ›²çº¿"""
    plt.figure(figsize=(10, 6))
    epochs = range(1, len(train_accuracies) + 1)

    plt.plot(epochs, train_accuracies, 'b-', label='è®­ç»ƒå‡†ç¡®ç‡', linewidth=2)
    plt.plot(epochs, val_accuracies, 'r-', label='éªŒè¯å‡†ç¡®ç‡', linewidth=2)

    plt.title('è®­ç»ƒå’ŒéªŒè¯å‡†ç¡®ç‡', fontsize=14, fontweight='bold')
    plt.xlabel('Epochs', fontsize=12)
    plt.ylabel('Accuracy (%)', fontsize=12)
    plt.legend(fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.ylim(bottom=0)

    # æ·»åŠ æœ€ä½³å‡†ç¡®ç‡æ ‡æ³¨
    best_val_acc = max(val_accuracies)
    best_epoch = val_accuracies.index(best_val_acc) + 1
    plt.axvline(x=best_epoch, color='gray', linestyle='--', alpha=0.7)
    plt.text(best_epoch, best_val_acc / 2, f'æœ€ä½³: {best_val_acc:.2f}%\nEpoch: {best_epoch}',
             ha='center', va='center', fontsize=10, 
             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        plt.close()
    else:
        plt.show()
```

```python
class SpeakerDataset(Dataset):
    def __init__(self, data_dir, segment_len=128):
        """åˆå§‹åŒ–è¯´è¯äººæ•°æ®é›†
        
        Args:
            data_dir (str): æ•°æ®ç›®å½•è·¯å¾„
            segment_len (int): æ¯ä¸ªè¯­éŸ³ç‰‡æ®µçš„å¸§æ•°ï¼Œé»˜è®¤ä¸º128
        """
        self.data_dir = data_dir
        self.segment_len = segment_len

        # åŠ è½½è¯´è¯äººæ˜ å°„å…³ç³»
        mapping_path = Path(data_dir) / "mapping.json"
        mapping = json.load(mapping_path.open())
        self.speaker2id = mapping["speaker2id"]

        # åŠ è½½å…ƒæ•°æ®
        metadata_path = Path(data_dir) / "metadata.json"
        metadata = json.load(open(metadata_path))["speakers"]

        # è·å–è¯´è¯äººæ€»æ•°
        self.speaker_num = len(metadata.keys())
        self.data = []
        
        # æ„å»ºæ•°æ®åˆ—è¡¨: [ç‰¹å¾è·¯å¾„, è¯´è¯äººID]
        for speaker in metadata.keys():
            for utterances in metadata[speaker]:
                self.data.append([utterances["feature_path"], self.speaker2id[speaker]])

    def __len__(self):
        """è¿”å›æ•°æ®é›†å¤§å°"""
        return len(self.data)

    def __getitem__(self, index):
        """è·å–å•ä¸ªæ ·æœ¬"""
        feat_path, speaker = self.data[index]
        
        # åŠ è½½melé¢‘è°±ç‰¹å¾
        mel = torch.load(os.path.join(self.data_dir, feat_path))

        # éšæœºæˆªå–å›ºå®šé•¿åº¦çš„ç‰‡æ®µ
        if len(mel) > self.segment_len:
            start = random.randint(0, len(mel) - self.segment_len)
            mel = torch.FloatTensor(mel[start:start+self.segment_len])
        else:
            mel = torch.FloatTensor(mel)
            
        speaker = torch.LongTensor([speaker]).squeeze()
        
        return mel, speaker

    def get_speaker_number(self):
        """è¿”å›è¯´è¯äººæ•°é‡"""
        return self.speaker_num
```

# dataloader

```python
import torch
from torch.utils.data import DataLoader, random_split
from torch.nn.utils.rnn import pad_sequence


def collate_batch(batch):
  """å¤„ç†ä¸€ä¸ªæ‰¹æ¬¡çš„æ•°æ®
  
  å°†åŒä¸€ä¸ªæ‰¹æ¬¡ä¸­çš„ç‰¹å¾è¿›è¡Œå¡«å……ï¼Œä½¿å®ƒä»¬çš„é•¿åº¦ç›¸åŒ
  
  Args:
      batch: ä¸€ä¸ªæ‰¹æ¬¡çš„æ•°æ®ï¼ŒåŒ…å«melé¢‘è°±å›¾å’Œè¯´è¯äººID
  
  Returns:
      tuple: å¡«å……åçš„melé¢‘è°±å›¾å’Œè¯´è¯äººIDå¼ é‡
  """
  # å°†æ‰¹æ¬¡æ•°æ®è§£åŒ…ä¸ºmelé¢‘è°±å›¾å’Œè¯´è¯äººID
  mel, speaker = zip(*batch)
  
  # å¯¹åŒä¸€ä¸ªæ‰¹æ¬¡ä¸­çš„melé¢‘è°±å›¾è¿›è¡Œå¡«å……ï¼Œä½¿å®ƒä»¬é•¿åº¦ç›¸åŒ
  # ä½¿ç”¨-20è¿›è¡Œå¡«å……ï¼Œå¯¹åº”log10^(-20)ï¼Œè¿™æ˜¯ä¸€ä¸ªéå¸¸å°çš„å€¼ï¼ˆæ¥è¿‘0ï¼‰
  mel = pad_sequence(mel, batch_first=True, padding_value=-20)
  
  # melçš„å½¢çŠ¶: (æ‰¹æ¬¡å¤§å°, åºåˆ—é•¿åº¦, 40ä¸ªmelé¢‘å¸¦)
  return mel, torch.FloatTensor(speaker).long()


def get_dataloader(data_dir, batch_size, n_workers):
  """ç”Ÿæˆæ•°æ®åŠ è½½å™¨
  
  Args:
      data_dir (str): æ•°æ®ç›®å½•è·¯å¾„
      batch_size (int): æ‰¹æ¬¡å¤§å°
      n_workers (int): æ•°æ®åŠ è½½çš„å·¥ä½œè¿›ç¨‹æ•°
  
  Returns:
      tuple: è®­ç»ƒæ•°æ®åŠ è½½å™¨ã€éªŒè¯æ•°æ®åŠ è½½å™¨ã€è¯´è¯äººæ•°é‡
  """
  # åˆ›å»ºæ•°æ®é›†å®ä¾‹
  dataset = SpeakerDataset(data_dir)
  
  # è·å–æ•°æ®é›†ä¸­è¯´è¯äººçš„æ€»æ•°
  speaker_num = dataset.get_speaker_number()
  
  # å°†æ•°æ®é›†æŒ‰9:1çš„æ¯”ä¾‹åˆ†å‰²ä¸ºè®­ç»ƒé›†å’ŒéªŒè¯é›†
  trainlen = int(0.9 * len(dataset))  # 90% è®­ç»ƒé›†
  lengths = [trainlen, len(dataset) - trainlen]  # è®­ç»ƒé›†å’ŒéªŒè¯é›†çš„å¤§å°
  trainset, validset = random_split(dataset, lengths)  # éšæœºåˆ†å‰²

  # åˆ›å»ºè®­ç»ƒæ•°æ®åŠ è½½å™¨
  train_loader = DataLoader(
    trainset,                    # è®­ç»ƒæ•°æ®é›†
    batch_size=batch_size,       # æ‰¹æ¬¡å¤§å°
    shuffle=True,                # æ¯ä¸ªepochæ‰“ä¹±æ•°æ®
    drop_last=True,              # ä¸¢å¼ƒæœ€åä¸€ä¸ªä¸å®Œæ•´çš„æ‰¹æ¬¡
    num_workers=n_workers,       # æ•°æ®åŠ è½½çš„å·¥ä½œè¿›ç¨‹æ•°
    pin_memory=True,             # å°†æ•°æ®å›ºå®šåœ¨å†…å­˜ä¸­ï¼ŒåŠ é€ŸGPUä¼ è¾“
    collate_fn=collate_batch,    # è‡ªå®šä¹‰æ‰¹æ¬¡å¤„ç†å‡½æ•°
  )
  
  # åˆ›å»ºéªŒè¯æ•°æ®åŠ è½½å™¨
  valid_loader = DataLoader(
    validset,                    # éªŒè¯æ•°æ®é›†
    batch_size=batch_size,       # æ‰¹æ¬¡å¤§å°
    num_workers=n_workers,       # æ•°æ®åŠ è½½çš„å·¥ä½œè¿›ç¨‹æ•°
    drop_last=True,              # ä¸¢å¼ƒæœ€åä¸€ä¸ªä¸å®Œæ•´çš„æ‰¹æ¬¡
    pin_memory=True,             # å°†æ•°æ®å›ºå®šåœ¨å†…å­˜ä¸­ï¼ŒåŠ é€ŸGPUä¼ è¾“
    collate_fn=collate_batch,    # è‡ªå®šä¹‰æ‰¹æ¬¡å¤„ç†å‡½æ•°
  )

  return train_loader, valid_loader, speaker_num
```

```python
import math
import torch
from torch.optim import Optimizer
from torch.optim.lr_scheduler import LambdaLR

def get_cosine_schedule_with_warmup(optimizer, num_warmup_steps, num_training_steps, num_cycles=0.5):
    """å¸¦çƒ­èº«é˜¶æ®µçš„ä½™å¼¦é€€ç«å­¦ä¹ ç‡è°ƒåº¦å™¨
    
    Args:
        optimizer: ä¼˜åŒ–å™¨
        num_warmup_steps: çƒ­èº«æ­¥æ•°
        num_training_steps: æ€»è®­ç»ƒæ­¥æ•°
        num_cycles: ä½™å¼¦å‘¨æœŸæ•°
        
    Returns:
        å­¦ä¹ ç‡è°ƒåº¦å™¨
    """
    def lr_lambda(current_step):
        if current_step < num_warmup_steps:
            # çº¿æ€§çƒ­èº«
            return float(current_step) / float(max(1, num_warmup_steps))
        else:
            # ä½™å¼¦é€€ç«
            progress = float(current_step - num_warmup_steps) / float(max(1, num_training_steps - num_warmup_steps))
            return max(0.0, 0.5 * (1.0 + math.cos(math.pi * float(num_cycles) * 2.0 * progress)))
    
    return LambdaLR(optimizer, lr_lambda)
```

```python
def model_fn(batch, model, criterion, device):
    """å•ä¸ªæ‰¹æ¬¡çš„å‰å‘ä¼ æ’­
    
    Args:
        batch: æ•°æ®æ‰¹æ¬¡
        model: æ¨¡å‹
        criterion: æŸå¤±å‡½æ•°
        device: è®¡ç®—è®¾å¤‡
        
    Returns:
        tuple: (æŸå¤±, å‡†ç¡®ç‡)
    """
    mels, labels = batch
    mels = mels.to(device)
    labels = labels.to(device)

    # å‰å‘ä¼ æ’­
    outputs = model(mels)
    
    # è®¡ç®—æŸå¤±å’Œå‡†ç¡®ç‡
    loss = criterion(outputs, labels)
    preds = outputs.argmax(dim=1)
    accuracy = (preds == labels).float().mean()
    
    return loss, accuracy
```

```python
def validate(dataloader, model, criterion, device):
    """åœ¨éªŒè¯é›†ä¸Šè¯„ä¼°æ¨¡å‹
    
    Args:
        dataloader: éªŒè¯æ•°æ®åŠ è½½å™¨
        model: æ¨¡å‹
        criterion: æŸå¤±å‡½æ•°
        device: è®¡ç®—è®¾å¤‡
        
    Returns:
        tuple: (å¹³å‡æŸå¤±, å¹³å‡å‡†ç¡®ç‡)
    """
    model.eval()
    total_loss = 0.0
    total_accuracy = 0.0
    num_batches = len(dataloader)
    
    with torch.no_grad():
        for batch in dataloader:
            loss, accuracy = model_fn(batch, model, criterion, device)
            total_loss += loss.item()
            total_accuracy += accuracy.item()
    
    model.train()
    
    return total_loss / num_batches, total_accuracy / num_batches
```

```python
# ç»˜åˆ¶è®­ç»ƒæ›²çº¿
print("å¼€å§‹ç»˜åˆ¶è®­ç»ƒæ›²çº¿...")
def plot_training_results(train_losses, val_losses, train_accuracies, val_accuracies, learning_rates, save_dir=None):
    """Plot complete training results charts
    
    Args:
        train_losses: List of training losses
        val_losses: List of validation losses
        train_accuracies: List of training accuracies
        val_accuracies: List of validation accuracies
        learning_rates: List of learning rates
        save_dir: Directory path to save plots
    """
    import matplotlib.pyplot as plt
    import numpy as np
    
    # Create subplots
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    # Calculate x-axis data
    epochs = range(1, len(val_losses) + 1)
    steps = range(1, len(train_losses) + 1)
    
    # 1. Training and Validation Loss (by epoch)
    ax1.plot(epochs, val_losses, 'r-', label='Validation Loss', linewidth=2, marker='o', markersize=4)
    
    # Calculate average training loss per epoch
    steps_per_epoch = len(train_losses) // len(val_losses)
    epoch_train_losses = []
    for i in range(len(val_losses)):
        start_idx = i * steps_per_epoch
        end_idx = min((i + 1) * steps_per_epoch, len(train_losses))
        epoch_avg_loss = np.mean(train_losses[start_idx:end_idx])
        epoch_train_losses.append(epoch_avg_loss)
    
    ax1.plot(epochs, epoch_train_losses, 'b-', label='Training Loss', linewidth=2, marker='s', markersize=4)
    ax1.set_title('Training and Validation Loss', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Mark best validation loss
    best_val_loss_idx = np.argmin(val_losses)
    best_val_loss = val_losses[best_val_loss_idx]
    ax1.axvline(x=best_val_loss_idx + 1, color='gray', linestyle='--', alpha=0.7)
    ax1.text(best_val_loss_idx + 1, best_val_loss, f'Best\n{best_val_loss:.4f}', 
             ha='center', va='bottom', fontsize=10,
             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
    
    # 2. Training and Validation Accuracy (by epoch)
    ax2.plot(epochs, val_accuracies, 'r-', label='Validation Accuracy', linewidth=2, marker='o', markersize=4)
    
    # Calculate average training accuracy per epoch
    epoch_train_accuracies = []
    for i in range(len(val_accuracies)):
        start_idx = i * steps_per_epoch
        end_idx = min((i + 1) * steps_per_epoch, len(train_accuracies))
        epoch_avg_acc = np.mean(train_accuracies[start_idx:end_idx])
        epoch_train_accuracies.append(epoch_avg_acc)
    
    ax2.plot(epochs, epoch_train_accuracies, 'b-', label='Training Accuracy', linewidth=2, marker='s', markersize=4)
    ax2.set_title('Training and Validation Accuracy', fontsize=14, fontweight='bold')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Mark best validation accuracy
    best_val_acc_idx = np.argmax(val_accuracies)
    best_val_acc = val_accuracies[best_val_acc_idx]
    ax2.axvline(x=best_val_acc_idx + 1, color='gray', linestyle='--', alpha=0.7)
    ax2.text(best_val_acc_idx + 1, best_val_acc, f'Best\n{best_val_acc:.4f}', 
             ha='center', va='bottom', fontsize=10,
             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
    
    # 3. Learning Rate Schedule (by step)
    ax3.plot(steps, learning_rates, 'g-', linewidth=2)
    ax3.set_title('Learning Rate Schedule', fontsize=14, fontweight='bold')
    ax3.set_xlabel('Step')
    ax3.set_ylabel('Learning Rate')
    ax3.set_yscale('log')
    ax3.grid(True, alpha=0.3)
    
    # 4. Detailed Training Process (by step)
    # Apply moving average to smooth training curves
    window_size = min(100, len(train_losses) // 10)
    if window_size > 1:
        smooth_train_losses = np.convolve(train_losses, np.ones(window_size)/window_size, mode='valid')
        smooth_train_accuracies = np.convolve(train_accuracies, np.ones(window_size)/window_size, mode='valid')
        smooth_steps = steps[window_size-1:]
        
        ax4.plot(smooth_steps, smooth_train_losses, 'b-', alpha=0.7, label='Training Loss (smoothed)')
        ax4_twin = ax4.twinx()
        ax4_twin.plot(smooth_steps, smooth_train_accuracies, 'r-', alpha=0.7, label='Training Accuracy (smoothed)')
    else:
        ax4.plot(steps, train_losses, 'b-', alpha=0.7, label='Training Loss')
        ax4_twin = ax4.twinx()
        ax4_twin.plot(steps, train_accuracies, 'r-', alpha=0.7, label='Training Accuracy')
    
    ax4.set_xlabel('Step')
    ax4.set_ylabel('Loss', color='blue')
    ax4_twin.set_ylabel('Accuracy', color='red')
    ax4.set_title('Training Process Details', fontsize=14, fontweight='bold')
    
    # Combine legends
    lines1, labels1 = ax4.get_legend_handles_labels()
    lines2, labels2 = ax4_twin.get_legend_handles_labels()
    ax4.legend(lines1 + lines2, labels1 + labels2, loc='upper right')
    
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    # Save plot
    if save_dir:
        plot_path = os.path.join(save_dir, "training_analysis.png")
        plt.savefig(plot_path, dpi=300, bbox_inches='tight')
        print(f"Training analysis plot saved: {plot_path}")
    
    plt.show()
    
    return fig
```
